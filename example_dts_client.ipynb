# Demo of the DTS validator client

```python
import logging
import sys
# by enabling the logging, we can see the HTTP requests made by the client
logging.basicConfig(stream=sys.stdout, level=logging.INFO)

from dts_validator.client import DTS_API
```

## DraCor API

```python
DTSDRACOR_DTS_API = "https://staging.dracor.org/api/v1/dts"
```

```python
dts_client = DTS_API(
    DTSDRACOR_DTS_API,
    enable_validation=False
)
```

```python
dts_client.is_validation_enabled
```

```python
# list all top collections
dracor_collections = dts_client.collections(recursive=False)
```

```python
len(dracor_collections)
```

```python
# the first 5 collections
dracor_collections[:5]
```

```python
# let's dig into the first collection
one_collection = dracor_collections[0]
```

```python
print(one_collection)
```

```python
from dts_validator.client import get_resource_recursively
```

```python
# TODO: make it really recursive
r = get_resource_recursively(one_collection, dts_client)
```

```python
# let's get all the resources (=documents) in this collection
resources = dts_client.collections(id=one_collection.id).children
```

```python
# how many resources are in this collection?
len(resources)
```

```python
resources[:5]
```

```python
one_resource = resources[0]
```

```python
one_resource
```

```python
one_resource.json
```

```python
navigation_obj, _ = dts_client.navigation(resource=one_resource, down=1)
```

```python
# if we want to know in which format the resource is available
# we can check the mediaTypes attribute of the resource
navigation_obj._json['resource']['mediaTypes']
```

```python
# what are the citable units at the first level of the resource?
navigation_obj.citable_units
```

```python
document_string, _ = dts_client.document(resource=one_resource)
```

```python
# as we haven't specified the format of the document
# we need to look into the response's headers to see in which format the document was returned
_.headers['Content-Type']
```

```python
print(document_string[:500])
```
